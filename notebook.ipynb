## Visualization & Explanatory Analysis

### Pertanyaan 1:  Bagaimana perubahan cuaca dapat mempengaruhi jumlah penyewaan sepeda?

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='weathersit', y='cnt', data=day_data)
plt.title('Effect of Weather on Bike Rentals')
plt.show()
```

### Pertanyaan 2: Bagaimana perubahan musim mempengaruhi jumlah penyewaan sepeda?

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='season', y='cnt', data=day_data)
plt.title('Bike Rentals Across Seasons')
plt.show()
```

**Insight:**
1. Pengaruh Cuaca terhadap Jumlah Penyewaan Sepeda:

  a) Tren Umum:
  
  Cuaca cerah cenderung menghasilkan jumlah penyewaan tertinggi.
  Cuaca buruk (hujan lebat atau salju) biasanya mengakibatkan penurunan drastis dalam penyewaan.

  b) Variabilitas:
  
  Cuaca cerah mungkin menunjukkan rentang data (box) yang lebih lebar, menandakan variabilitas tinggi dalam jumlah penyewaan.
  Cuaca buruk kemungkinan memiliki rentang data yang lebih sempit, menunjukkan konsistensi dalam penurunan penyewaan.

  c) Outlier:
  
  Mungkin ada beberapa outlier pada hari-hari dengan cuaca baik, menunjukkan kejadian khusus atau hari libur yang meningkatkan penyewaan secara signifikan.
  
  d) Implikasi Bisnis:
  
  Strategi promosi bisa difokuskan pada hari-hari dengan prakiraan cuaca baik.
  Perlu ada rencana mitigasi untuk mempertahankan pendapatan pada hari-hari dengan cuaca buruk, seperti penawaran khusus atau layanan alternatif.


2. Perubahan Jumlah Penyewaan Sepeda Berdasarkan Musim:

  a) Pola Musiman:
  
  Musim panas dan musim gugur kemungkinan menunjukkan jumlah penyewaan tertinggi.
  Musim dingin mungkin memiliki jumlah penyewaan terendah karena cuaca yang tidak mendukung.

  b) Konsistensi vs Variabilitas:
  
  Musim semi dan musim gugur mungkin menunjukkan variabilitas yang lebih tinggi (box plot lebih lebar) karena perubahan cuaca yang lebih dinamis.
  Musim panas mungkin memiliki median tertinggi dengan variabilitas moderat.

  c) Tren Antar Musim:
  
  Mungkin terlihat peningkatan bertahap dari musim dingin ke musim semi, puncak di musim panas, dan sedikit penurunan di musim gugur.
  
  d) Implikasi Operasional:
  
  Kebutuhan untuk menyesuaikan inventaris dan staf berdasarkan perubahan musiman.
  Peluang untuk strategi pemasaran dan promosi yang disesuaikan dengan musim.
